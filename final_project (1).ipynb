```python
#preprocessing
#1 resizing
#2 flatten

import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize

target = []
images = []
flat_data = []

DATADIR = '/content/images'
CATEGORIES = ['Alluvial Soil', 'black soil', 'forest soil', 'red soil']
for category in CATEGORIES:
    class_num = CATEGORIES.index(category)  # label encoding the values
    path = os.path.join(DATADIR, category)  # create a path to use all the image
    for img in os.listdir(path):
      img_array = imread(os.path.join(path, img))
      # print(img_array.shape)
      #plt.imshow(img_array)
      img_resized=resize(img_array,(150,150,3))
      flat_data.append(img_resized)
      target.append(class_num)
flat_data=np.array(flat_data)
target=np.array(target)
image=np.array(images)









```

```python
unique,count=np.unique(target,return_counts=True)
plt.bar(CATEGORIES,count)
```

```python
#split data into Training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flat_data,target,test_size=0.3,random_state=109)
```

```python
!pip install pandas
```

```python
print("x_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
```

```python
# Flatten each image in x_train
num_samples, height, width, channels = x_train.shape
x_train_flattened = x_train.reshape(num_samples, height * width * channels)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = [
    {'C': [1, 10, 100, 1000], 'kernel': ['linear']},
    {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel': ['rbf']},
]
svc = svm.SVC(probability=True)
clf = GridSearchCV(svc, param_grid)
clf.fit(x_train_flattened, y_train)

```

```python
# %%writefile app.py
# import streamlit as st
# import numpy as np
# from skimage.io import imread
# from skimage.transform import resize
# import pickle
# from PIL import Image
# import matplotlib.pyplot as plt

# st.set_page_config(
#     page_title="Soil Analysis",
#     page_icon=":herb:"
# )

# # Load trained model
# model = pickle.load(open('img_model.p', 'rb'))

# # CSS styling for file uploader
# file_uploader_style = """
# <style>
# #file_uploader {
#     border: 2px dashed #008080;
#     border-radius: 5px;
#     padding: 20px;
#     text-align: center;
#     font-size: 18px;
#     color: #008080;
#     background-color: #f0f0f0;
#     position: relative;
# }

# #file_uploader input[type="file"] {
#     position: absolute;
#     top: 0;
#     left: 0;
#     width: 100%;
#     height: 100%;
#     opacity: 0;
#     cursor: pointer;
# }

# #file_uploader::before {
#     content: "+";
#     position: absolute;
#     top: 50%;
#     left: 50%;
#     transform: translate(-50%, -50%);
#     font-size: 32px;
#     color: #008080;
# }

# #file_uploader:hover::before {
#     color: #005a5a;
# }
# </style>
# """

# st.markdown(
#     """
#     <h1 style='text-align: center; color: #008080;'>Soil Analysis</h1>
#     <p style='text-align: center; font-size: 18px;'>Upload an image of soil to predict its type.</p>
#     """ + file_uploader_style,
#     unsafe_allow_html=True
# )

# uploaded_file = st.file_uploader(" ", type="jpg", help="Supported formats: jpg", key="file_uploader")

# if uploaded_file is not None:
#     img = Image.open(uploaded_file)
#     st.image(img, caption='Uploaded Image', use_column_width=True)

#     if st.button('PREDICT'):
#         CATEGORIES = ['Alluvial Soil', 'Black Soil', 'Forest Soil', 'Red Soil']
#         st.write('Result...')
#         flat_data = []
#         img = np.array(img)
#         img_resized = resize(img, (150, 150, 3))
#         flat_data.append(img_resized.flatten())
#         flat_data = np.array(flat_data)

#         y_out = model.predict(flat_data)
#         predicted_label = CATEGORIES[y_out[0]]
#         st.success(f'PREDICTED OUTPUT: {predicted_label}')

#         fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
#         ax1.imshow(img)
#         ax1.set_title('Original Image')
#         ax1.axis('off')

#         ax2.imshow(img)
#         ax2.set_title(f'Predicted: {predicted_label}')
#         #ax2.text(0, -20, f'Predicted: {predicted_label}', color='red', fontsize=12, bbox=dict(facecolor='white', alpha=0.7))
#         ax2.axis('off')

#         st.pyplot(fig)
#         # Assuming you have a dictionary mapping soil types to crops
# soil_to_crops = {
#     'Alluvial Soil': ['Rice', 'Wheat', 'Maize'],
#     'Black Soil': ['Cotton', 'Soybean', 'Groundnut'],
#     'Forest Soil': ['Tea', 'Coffee', 'Spices'],
#     'Red Soil': ['Tomato', 'Chili', 'Brinjal']
# }

# # After predicting soil type
# predicted_label = CATEGORIES[y_out[0]]
# suggested_crops = soil_to_crops.get(predicted_label, ['No suggestions found'])

# # Display predicted label and suggested crops
# st.success(f'PREDICTED OUTPUT: {predicted_label}')
# st.info(f'Suggested Crops: {", ".join(suggested_crops)}')
%%writefile app.py
import streamlit as st
import numpy as np
from skimage.io import imread
from skimage.transform import resize
import pickle
from PIL import Image
import matplotlib.pyplot as plt

st.set_page_config(
    page_title="Soil Analysis",
    page_icon=":herb:"
)

# Load trained model
model = pickle.load(open('img_model.p', 'rb'))

# CSS styling for file uploader
file_uploader_style = """
<style>
#file_uploader {
    border: 2px dashed #008080;
    border-radius: 5px;
    padding: 20px;
    text-align: center;
    font-size: 18px;
    color: #008080;
    background-color: #f0f0f0;
    position: relative;
}

#file_uploader input[type="file"] {
    position: absolute;
    top: 0;
    left: 0;
    width: 100%;
    height: 100%;
    opacity: 0;
    cursor: pointer;
}

#file_uploader::before {
    content: "+";
    position: absolute;
    top: 50%;
    left: 50%;
    transform: translate(-50%, -50%);
    font-size: 32px;
    color: #008080;
}

#file_uploader:hover::before {
    color: #005a5a;
}
</style>
"""

st.markdown(
    """
    <h1 style='text-align: center; color: #008080;'>Soil Analysis</h1>
    <p style='text-align: center; font-size: 18px;'>Upload an image of soil to predict its type.</p>
    """ + file_uploader_style,
    unsafe_allow_html=True
)

uploaded_file = st.file_uploader(" ", type="jpg", help="Supported formats: jpg", key="file_uploader")

if uploaded_file is not None:
    img = Image.open(uploaded_file)
    st.image(img, caption='Uploaded Image', use_column_width=True)

    if st.button('PREDICT'):
        CATEGORIES = ['Alluvial Soil', 'Black Soil', 'Forest Soil', 'Red Soil']
        st.write('Result...')
        flat_data = []
        img = np.array(img)
        img_resized = resize(img, (150, 150, 3))
        flat_data.append(img_resized.flatten())
        flat_data = np.array(flat_data)

        y_out = model.predict(flat_data)
        predicted_label = CATEGORIES[y_out[0]]
        st.success(f'PREDICTED OUTPUT: {predicted_label}')

        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
        ax1.imshow(img)
        ax1.set_title('Original Image')
        ax1.axis('off')

        ax2.imshow(img)
        ax2.set_title(f'Predicted: {predicted_label}')
        #ax2.text(0, -20, f'Predicted: {predicted_label}', color='red', fontsize=12, bbox=dict(facecolor='white', alpha=0.7))
        ax2.axis('off')

        st.pyplot(fig)

# Define soil_to_crops dictionary
soil_to_crops = {
    'Alluvial Soil': ['Rice', 'Wheat', 'Maize'],
    'Black Soil': ['Cotton', 'Soybean', 'Groundnut'],
    'Forest Soil': ['Tea', 'Coffee', 'Spices'],
    'Red Soil': ['Tomato', 'Chili', 'Brinjal']
}

# Display suggested crops after predicting soil type
if 'y_out' in locals():  # Check if y_out is defined
    predicted_label = CATEGORIES[y_out[0]]
    suggested_crops = soil_to_crops.get(predicted_label, ['No suggestions found'])
    st.success(f'PREDICTED OUTPUT: {predicted_label}')
    st.info(f'Suggested Crops: {", ".join(suggested_crops)}')

```
